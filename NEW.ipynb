```python
from sentence_transformers import util
import pandas as pd
import numpy as np
from sentence_transformers import InputExample

def load_examples_from_csv(csv_path):
    df = pd.read_csv(csv_path, on_bad_lines='skip')
    
    # Debug: print column names
    print("Columns found:", df.columns.tolist())

    # Rename if needed
    expected_columns = ['word1','word2','label']
    if not all(col in df.columns for col in expected_columns):
        df.rename(columns={
            'Word A': 'word1',
            ' word2':'word2',
            ' label':'label', 
        }, inplace=True)

    # Drop any remaining bad rows
    df = df.dropna(subset=['word1', 'word2', 'label'])

    # Convert to InputExamples
    examples = [
        InputExample(texts=[str(row['word1']), str(row['word2'])], label=np.float32(row['label']))
        for _, row in df.iterrows()
    ]
    return examples
```

```python
import csv
with open('GFG.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        print(''.join(row))
```

```python
train_examples = load_examples_from_csv("GFG.csv")

# Model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Dataloader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=4)

# Loss
train_loss = losses.CosineSimilarityLoss(model=model)

# Training
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=5,
    warmup_steps=10,
    show_progress_bar=True
)

# Save
model.save("custom-word-similarity-model")
```
